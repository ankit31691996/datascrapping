```python
from bs4 import BeautifulSoup
import requests
```

```python
bowling1= requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/bowling')
bowling1
```

```python
bowling=BeautifulSoup(bowling1.content)
bowling
```

```python
name=[]
name1=bowling.find('div',class_='rankings-block__banner--name-large')
name.append(name1.text)
name
```

```python
ratings=[]
rating1=bowling.find('div',class_='rankings-block__banner--rating')
ratings.append(rating1.text)
ratings
```

```python
country=[]
country1=bowling.find('div',class_='rankings-block__banner--nationality')
country.append(country1.text.replace('\n',''))
country
```

```python
career_best=[]
career=bowling.find('span',class_='rankings-block__career-best-text')
career_best.append(career.text.replace('\n',''))
career_best
```

```python
for i in bowling.find_all('td',class_='table-body__cell rankings-table__name name'):
    name.append(i.text.replace('\n',''))
name
```

```python
for i in bowling.find_all('span',class_='table-body__logo-text'):
    country.append(i.text)
country
```

```python
for i in bowling.find_all('td',class_='table-body__cell rating'):
    ratings.append(i.text)
ratings
```

```python
for i in bowling.find_all('td',class_='table-body__cell u-text-right u-hide-phablet'):
    career_best.append(i.text.replace('\n',''))
career_best
```

```python
print(len(name[0:10]),len(country[0:10]),len(ratings[0:10]),len(career_best[0:10]))
```

```python
import pandas as pd
df=pd.DataFrame({"Player Name":name[0:10],"Country":country[0:10],'Ratings':ratings[0:10],"Career Best":career_best[0:10]})
df
```
