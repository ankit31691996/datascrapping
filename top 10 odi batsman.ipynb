```python
from bs4 import BeautifulSoup
import requests
```

```python
batsmen= requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
batsmen
```

```python
batsmen1=BeautifulSoup(batsmen.content)
batsmen1
```

```python
name=[]
name1=batsmen1.find('div',class_='rankings-block__banner--name-large')
name1.text
```

```python
name=[name1.text]
```

```python
name
```

```python
country=[]
country1=batsmen1.find('div',class_='rankings-block__banner--nationality')
country.append(country1.text.replace('\n',''))
country
```

```python
ratings=[]
rating1=batsmen1.find('div',class_='rankings-block__banner--rating')
ratings.append(rating1.text)
ratings
```

```python
career_best=[]
career=batsmen1.find('span',class_='rankings-block__career-best-text')
career_best.append(career.text.replace('\n',''))
career_best
```

```python
for i in batsmen1.find_all('td',class_='table-body__cell rankings-table__name name'):
    name.append(i.text.replace('\n',''))
name
```

```python
for i in batsmen1.find_all('span',class_='table-body__logo-text'):
    country.append(i.text)
country
```

```python
for i in batsmen1.find_all('td',class_='table-body__cell rating'):
    ratings.append(i.text)
ratings
```

```python
for i in batsmen1.find_all('td',class_='table-body__cell u-text-right u-hide-phablet'):
    career_best.append(i.text.replace('\n',''))
career_best
```

```python
print(len(name[0:10]),len(country[0:10]),len(ratings[0:10]),len(career_best[0:10]))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"Player Name":name[0:10],"Country":country[0:10],'Ratings':ratings[0:10],"Career Best":career_best[0:10]})
df
```
